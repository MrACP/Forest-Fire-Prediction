Creating class 'Forest fire'

```python
class Forest_fire():
    temp = int(input("Temperature in ° Celsius: "))                     #Asuming the range of 2°C to 40°C
    wind_speed = int(input("Enter wind speed in mph: "))                #Asuming the range of 1 mph to 10 mph
    humidity = int(input("Is there Humidity present in the atmosphere? Yes [1] OR No [0]: "))

    def predict(self):
        if self.temp >= 30 and self.wind_speed >= 5:
            if self.humidity == 0:
                print("There are strong chances of Forest fire")
            else:
                print("There are some chances of Forest fire")
        else:
            print("There won't be any chances of Forest fire")

chance = Forest_fire()
chance.predict()
```

Creating a class 'Weather"

```python
class Weather():
    def outcome(self,weather):
      self.weather=weather

      if weather == 'Monsoon':
        print("There are less chances of fire.")
      elif weather=='Summer':
        print('There are high chances of fire.')
      else:
        print('There might be chances of fire.')
w1=Weather()
w1.outcome('Monsoon')
```
